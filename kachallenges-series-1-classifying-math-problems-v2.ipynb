```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split, StratifiedKFold
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
from transformers import BertTokenizer, BertForSequenceClassification, RobertaTokenizer, RobertaForSequenceClassification
from tqdm.notebook import tqdm
import random
import time
import gc
from scipy import stats

# Set random seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed(42)

# Define paths
DATA_PATH = "/kaggle/input/classification-of-math-problems-by-kasut-academy"
TRAIN_PATH = os.path.join(DATA_PATH, "train.csv")
TEST_PATH = os.path.join(DATA_PATH, "test.csv")
SAMPLE_SUBMISSION_PATH = os.path.join(DATA_PATH, "sample_submission.csv")
OUTPUT_DIR = "/kaggle/working/model_outputs"

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load data
print("Loading data...")
train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)
sample_submission = pd.read_csv(SAMPLE_SUBMISSION_PATH)

print(f"Train data shape: {train_df.shape}")
print(f"Test data shape: {test_df.shape}")

# Display class distribution
print("\nClass distribution:")
class_distribution = train_df["label"].value_counts().sort_index()
print(class_distribution)

# Visualize class distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=train_df["label"])
plt.title("Class Distribution in Training Data")
plt.xlabel("Topic Label")
plt.ylabel("Count")
plt.xticks(ticks=range(8), labels=[
    "Algebra", 
    "Geometry and Trigonometry", 
    "Calculus and Analysis",
    "Probability and Statistics",
    "Number Theory",
    "Combinatorics and Discrete Math",
    "Linear Algebra",
    "Abstract Algebra and Topology"
], rotation=45, ha="right")
plt.tight_layout()
plt.savefig("class_distribution.png")
plt.show()

# Define target names for reference
target_names = [
    "Algebra", 
    "Geometry and Trigonometry", 
    "Calculus and Analysis",
    "Probability and Statistics",
    "Number Theory",
    "Combinatorics and Discrete Math",
    "Linear Algebra",
    "Abstract Algebra and Topology"
]

# Define dataset class
class MathDataset(Dataset):
    def __init__(self, texts, labels=None, tokenizer=None, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=True,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        item = {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
        }
        
        if 'token_type_ids' in encoding:
            item['token_type_ids'] = encoding['token_type_ids'].flatten()
            
        if self.labels is not None:
            item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
            
        return item

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nUsing device: {device}")

# Define training function
def train_epoch(model, data_loader, optimizer, device, scheduler=None):
    model.train()
    total_loss = 0
    all_preds = []
    all_labels = []
    
    for batch in tqdm(data_loader, desc="Training"):
        optimizer.zero_grad()
        
        # Move batch to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        
        # Prepare input dict based on model type
        inputs = {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'labels': labels
        }
        
        if 'token_type_ids' in batch:
            inputs['token_type_ids'] = batch['token_type_ids'].to(device)
        
        # Forward pass
        outputs = model(**inputs)
        
        loss = outputs.loss
        total_loss += loss.item()
        
        # Get predictions
        preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())
        
        # Backward pass
        loss.backward()
        
        # Apply gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        if scheduler:
            scheduler.step()
    
    # Calculate metrics
    f1 = f1_score(all_labels, all_preds, average='micro')
    avg_loss = total_loss / len(data_loader)
    
    return avg_loss, f1

# Define evaluation function
def evaluate(model, data_loader, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []
    all_logits = []
    
    with torch.no_grad():
        for batch in tqdm(data_loader, desc="Evaluating"):
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            
            # Prepare input dict based on model type
            inputs = {
                'input_ids': input_ids,
                'attention_mask': attention_mask,
                'labels': labels
            }
            
            if 'token_type_ids' in batch:
                inputs['token_type_ids'] = batch['token_type_ids'].to(device)
            
            # Forward pass
            outputs = model(**inputs)
            
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions
            logits = outputs.logits
            all_logits.append(logits.cpu())
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())
    
    # Calculate metrics
    f1_micro = f1_score(all_labels, all_preds, average='micro')
    f1_macro = f1_score(all_labels, all_preds, average='macro')
    f1_weighted = f1_score(all_labels, all_preds, average='weighted')
    
    # Stack all logits
    all_logits = torch.cat(all_logits, dim=0)
    
    results = {
        'loss': total_loss / len(data_loader),
        'f1_micro': f1_micro,
        'f1_macro': f1_macro,
        'f1_weighted': f1_weighted,
        'preds': all_preds,
        'labels': all_labels,
        'logits': all_logits
    }
    
    return results

# Define prediction function
def predict(model, data_loader, device, return_logits=False):
    model.eval()
    all_preds = []
    all_logits = []
    
    with torch.no_grad():
        for batch in tqdm(data_loader, desc="Predicting"):
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            
            # Prepare input dict based on model type
            inputs = {
                'input_ids': input_ids,
                'attention_mask': attention_mask
            }
            
            if 'token_type_ids' in batch:
                inputs['token_type_ids'] = batch['token_type_ids'].to(device)
            
            # Forward pass
            outputs = model(**inputs)
            
            # Get predictions
            logits = outputs.logits
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            
            if return_logits:
                all_logits.append(logits.cpu())
    
    if return_logits:
        return all_preds, torch.cat(all_logits, dim=0)
    else:
        return all_preds

# Function to train a model
def train_model(model, train_loader, val_loader, optimizer, scheduler, device, n_epochs=3, early_stopping_patience=2, model_name="model"):
    best_f1 = 0
    best_model_state = None
    patience_counter = 0
    
    train_losses = []
    train_f1s = []
    val_losses = []
    val_f1s = []
    
    for epoch in range(n_epochs):
        print(f"Epoch {epoch+1}/{n_epochs}")
        
        # Train
        train_loss, train_f1 = train_epoch(model, train_loader, optimizer, device, scheduler)
        train_losses.append(train_loss)
        train_f1s.append(train_f1)
        print(f"Train Loss: {train_loss:.4f}, Train F1: {train_f1:.4f}")
        
        # Evaluate
        val_results = evaluate(model, val_loader, device)
        val_loss = val_results['loss']
        val_f1 = val_results['f1_micro']
        val_losses.append(val_loss)
        val_f1s.append(val_f1)
        print(f"Val Loss: {val_loss:.4f}, Val F1 Micro: {val_f1:.4f}")
        
        # Save best model
        if val_f1 > best_f1:
            best_f1 = val_f1
            best_model_state = model.state_dict().copy()
            patience_counter = 0
            print(f"New best model saved with F1: {best_f1:.4f}")
            
            # Save model checkpoint
            torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, f"best_{model_name}.pt"))
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print(f"Early stopping triggered after {epoch+1} epochs")
                break
    
    # Plot training curves
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title("Loss Curves")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(train_f1s, label="Train F1")
    plt.plot(val_f1s, label="Val F1")
    plt.title("F1 Curves")
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.legend()
    
    plt.tight_layout()
    plt.savefig(f"{model_name}_training_curves.png")
    plt.show()
    
    return best_model_state, best_f1

# Function to implement k-fold cross-validation with model ensemble
def train_ensemble(data, models_config, n_folds=5, seed=42):
    # Define model configurations
    models_info = []
    
    # Initialize stratified k-fold
    skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=seed)
    
    # Model weights for ensemble (will be updated based on validation performance)
    model_weights = np.ones(len(models_config)) / len(models_config)
    
    # Store fold models for ensemble prediction
    fold_models = []
    
    # Iterate through folds
    for fold, (train_idx, val_idx) in enumerate(skf.split(data['Question'], data['label'])):
        print(f"\n=========== Fold {fold+1}/{n_folds} ===========")
        
        # Split data
        fold_train_data = data.iloc[train_idx].reset_index(drop=True)
        fold_val_data = data.iloc[val_idx].reset_index(drop=True)
        
        print(f"Training size: {len(fold_train_data)}, Validation size: {len(fold_val_data)}")
        
        fold_models_states = []
        fold_f1_scores = []
        
        # Train each model type on this fold
        for model_idx, model_cfg in enumerate(models_config):
            print(f"\nTraining {model_cfg['name']} model for fold {fold+1}")
            
            # Load tokenizer
            print(f"Loading tokenizer: {model_cfg['model_name']}")
            if "roberta" in model_cfg['model_name'].lower():
                tokenizer = RobertaTokenizer.from_pretrained(model_cfg['model_name'])
            else:
                tokenizer = BertTokenizer.from_pretrained(model_cfg['model_name'])
            
            # Prepare datasets
            fold_train_dataset = MathDataset(
                fold_train_data['Question'].values, 
                fold_train_data['label'].values, 
                tokenizer,
                max_len=model_cfg['max_len']
            )
            
            fold_val_dataset = MathDataset(
                fold_val_data['Question'].values, 
                fold_val_data['label'].values, 
                tokenizer,
                max_len=model_cfg['max_len']
            )
            
            # Create data loaders
            fold_train_loader = DataLoader(fold_train_dataset, batch_size=model_cfg['batch_size'], shuffle=True)
            fold_val_loader = DataLoader(fold_val_dataset, batch_size=model_cfg['batch_size']*2, shuffle=False)
            
            # Load model
            print(f"Loading model: {model_cfg['model_name']}")
            if "roberta" in model_cfg['model_name'].lower():
                model = RobertaForSequenceClassification.from_pretrained(
                    model_cfg['model_name'],
                    num_labels=8
                )
            else:
                model = BertForSequenceClassification.from_pretrained(
                    model_cfg['model_name'],
                    num_labels=8
                )
            model.to(device)
            
            # Define optimizer
            optimizer = optim.AdamW(
                model.parameters(), 
                lr=model_cfg['learning_rate'], 
                weight_decay=model_cfg['weight_decay']
            )
            
            # Define learning rate scheduler
            total_steps = len(fold_train_loader) * model_cfg['epochs']
            scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=total_steps)
            
            # Train model
            best_model_state, best_f1 = train_model(
                model, 
                fold_train_loader, 
                fold_val_loader, 
                optimizer, 
                scheduler, 
                device, 
                n_epochs=model_cfg['epochs'], 
                early_stopping_patience=model_cfg['patience'],
                model_name=f"{model_cfg['name']}_fold{fold+1}"
            )
            
            # Store best model state and F1 score
            fold_models_states.append(best_model_state)
            fold_f1_scores.append(best_f1)
            
            # Save memory
            del model, optimizer, scheduler, fold_train_dataset, fold_val_dataset
            del fold_train_loader, fold_val_loader
            gc.collect()
            torch.cuda.empty_cache()
        
        # Store fold models and their weights
        fold_models.append({
            'models_config': models_config,
            'model_states': fold_models_states,
            'f1_scores': fold_f1_scores
        })
        
        # Update global model weights based on validation performance
        fold_weights = np.array(fold_f1_scores) / np.sum(fold_f1_scores)
        model_weights = (model_weights + fold_weights) / 2
    
    # Normalize weights
    model_weights = model_weights / np.sum(model_weights)
    
    print("\n===== Ensemble Model Weights =====")
    for i, (weight, model_cfg) in enumerate(zip(model_weights, models_config)):
        print(f"{model_cfg['name']}: {weight:.4f}")
    
    return fold_models, model_weights

# Function to make ensemble predictions
def ensemble_predict(test_data, fold_models, model_weights):
    all_fold_predictions = []
    
    # For each fold's models
    for fold_idx, fold_data in enumerate(fold_models):
        fold_predictions = np.zeros((len(test_data), 8))  # 8 classes
        
        # For each model in the fold
        for model_idx, (model_cfg, model_state) in enumerate(zip(fold_data['models_config'], fold_data['model_states'])):
            print(f"\nPredicting with {model_cfg['name']} model from fold {fold_idx+1}")
            
            # Load tokenizer
            if "roberta" in model_cfg['model_name'].lower():
                tokenizer = RobertaTokenizer.from_pretrained(model_cfg['model_name'])
                model = RobertaForSequenceClassification.from_pretrained(
                    model_cfg['model_name'],
                    num_labels=8
                )
            else:
                tokenizer = BertTokenizer.from_pretrained(model_cfg['model_name'])
                model = BertForSequenceClassification.from_pretrained(
                    model_cfg['model_name'],
                    num_labels=8
                )
            
            # Load model weights
            model.load_state_dict(model_state)
            model.to(device)
            
            # Prepare dataset and dataloader
            test_dataset = MathDataset(
                test_data['Question'].values, 
                None, 
                tokenizer,
                max_len=model_cfg['max_len']
            )
            test_loader = DataLoader(test_dataset, batch_size=model_cfg['batch_size']*2, shuffle=False)
            
            # Get predictions and logits
            _, logits = predict(model, test_loader, device, return_logits=True)
            
            # Apply softmax to get probabilities
            probabilities = torch.nn.functional.softmax(logits, dim=1).numpy()
            
            # Weight predictions by model's weight
            model_weight = model_weights[model_idx]
            fold_predictions += probabilities * model_weight
            
            # Clean up to save memory
            del model, tokenizer, test_dataset, test_loader
            gc.collect()
            torch.cuda.empty_cache()
        
        all_fold_predictions.append(fold_predictions)
    
    # Average predictions across folds
    final_predictions = np.mean(all_fold_predictions, axis=0)
    
    # Get predicted class
    predicted_classes = np.argmax(final_predictions, axis=1)
    
    return predicted_classes, final_predictions

# Define model configurations for ensemble
models_config = [
    {
        'name': 'SciBERT',
        'model_name': 'allenai/scibert_scivocab_uncased',
        'max_len': 128,
        'batch_size': 16,
        'learning_rate': 2e-5,
        'weight_decay': 0.01,
        'epochs': 4,
        'patience': 2
    },
    {
        'name': 'RoBERTa-Base',
        'model_name': 'roberta-base',
        'max_len': 128,
        'batch_size': 16,
        'learning_rate': 1e-5,
        'weight_decay': 0.01,
        'epochs': 3,
        'patience': 2
    },
    {
        'name': 'BERT-Base',
        'model_name': 'bert-base-uncased',
        'max_len': 128,
        'batch_size': 16,
        'learning_rate': 3e-5,
        'weight_decay': 0.01,
        'epochs': 3,
        'patience': 2
    }
]

# Split train data into train and validation sets
train_data, val_data = train_test_split(
    train_df, 
    test_size=0.1, 
    stratify=train_df["label"],
    random_state=42
)

print(f"\nTraining data size: {train_data.shape[0]}")
print(f"Validation data size: {val_data.shape[0]}")

# Train the ensemble model
print("\n========== Training Ensemble Model ==========")
train_start = time.time()

fold_models, model_weights = train_ensemble(
    train_df,  # Use all training data for k-fold
    models_config,
    n_folds=3  # Reduced for efficiency, can use 5 for better results
)

train_end = time.time()
print(f"\nEnsemble training completed in {(train_end - train_start) / 60:.2f} minutes")

# Make ensemble predictions on validation data
print("\n========== Evaluating Ensemble on Validation Set ==========")
val_pred_classes, val_pred_probs = ensemble_predict(val_data, fold_models, model_weights)

# Calculate validation metrics
val_f1_micro = f1_score(val_data['label'], val_pred_classes, average='micro')
val_f1_macro = f1_score(val_data['label'], val_pred_classes, average='macro')
val_f1_weighted = f1_score(val_data['label'], val_pred_classes, average='weighted')

print(f"Ensemble Validation F1 Micro: {val_f1_micro:.4f}")
print(f"Ensemble Validation F1 Macro: {val_f1_macro:.4f}")
print(f"Ensemble Validation F1 Weighted: {val_f1_weighted:.4f}")

# Generate confusion matrix
cm = confusion_matrix(val_data['label'], val_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (Ensemble)")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.xticks(ticks=np.arange(8) + 0.5, labels=range(8))
plt.yticks(ticks=np.arange(8) + 0.5, labels=range(8))
plt.tight_layout()
plt.savefig("ensemble_confusion_matrix.png")
plt.show()

# Print classification report
print("\nClassification Report (Ensemble):")
print(classification_report(val_data['label'], val_pred_classes, target_names=target_names))

# Make ensemble predictions on test data
print("\n========== Making Test Predictions with Ensemble ==========")
test_pred_classes, test_pred_probs = ensemble_predict(test_df, fold_models, model_weights)

# Create submission file
print("Creating submission file...")
submission = sample_submission.copy()
submission["label"] = test_pred_classes
submission.to_csv("ensemble_submission.csv", index=False)
print("Ensemble submission file created!")

# Print predicted class distribution
print("\nPredicted label distribution in test data:")
print(pd.Series(test_pred_classes).value_counts().sort_index())

# Plot test predictions distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=test_pred_classes)
plt.title("Predicted Class Distribution in Test Data")
plt.xlabel("Topic Label")
plt.ylabel("Count")
plt.xticks(ticks=range(8), labels=target_names, rotation=45, ha="right")
plt.tight_layout()
plt.savefig("test_predictions_distribution.png")
plt.show()
```
